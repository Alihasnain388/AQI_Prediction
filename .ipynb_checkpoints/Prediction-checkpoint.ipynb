```python
import os
import joblib
import mlflow
import mlflow.sklearn
import pandas as pd
from pymongo import MongoClient
from datetime import datetime, timedelta

# =========================================================
# 1. Map your custom ENV names to MLflow ENV names
# =========================================================
os.environ["MLFLOW_TRACKING_USERNAME"] = os.getenv("DAGSHUB_USERNAME", "")
os.environ["MLFLOW_TRACKING_PASSWORD"] = os.getenv("DAGSHUB_TOKEN", "")

# Set the tracking URI (No popup, fully serverless)
REPO_OWNER = "Alihasnain388"
REPO_NAME = "AQI_Model"
mlflow.set_tracking_uri(f"https://dagshub.com/{REPO_OWNER}/{REPO_NAME}.mlflow")

# =========================================================
# 2. Load Model & Scaler from MLflow Registry
# =========================================================
client = mlflow.tracking.MlflowClient()
MODEL_NAME = "Karachi_AQI_Model"

try:
    # Get latest version
    versions = client.search_model_versions(filter_string=f"name='{MODEL_NAME}'")
    latest_version_info = sorted(versions, key=lambda x: int(x.version), reverse=True)[0]
    run_id = latest_version_info.run_id
    version_num = latest_version_info.version
    
    print(f"âœ… Using Model Version: {version_num}")

    # Load Model directly
    model_uri = f"models:/{MODEL_NAME}/{version_num}"
    model = mlflow.sklearn.load_model(model_uri)

    # Download and load Scaler artifact
    scaler_path = mlflow.artifacts.download_artifacts(
        run_id=run_id, 
        artifact_path="model/scaler.pkl"
    )
    scaler = joblib.load(scaler_path)
    print("âœ… Model and Scaler loaded from DagsHub")

except Exception as e:
    print(f"âŒ Error during model retrieval: {e}")
    exit(1)

# =========================================================
# 3. Fetch Latest Features from MongoDB
# =========================================================
mongo_uri = os.getenv("MONGO_URI")
if not mongo_uri:
    raise ValueError("âŒ MONGO_URI environment variable is not set")

mongo_client = MongoClient(mongo_uri)
db = mongo_client["AQIPredictionSystem"]
latest_data = db["karachi_features"].find_one(sort=[("timestamp", -1)])

if latest_data is None:
    raise ValueError("âŒ No feature data found in MongoDB")

# =========================================================
# 4. Prediction Logic
# =========================================================
feature_columns = [
    "hour", "day_of_week", "aqi_lag_1h", 
    "aqi_lag_24h", "aqi_change_rate", 
    "Wind_Speed_kmh", "PM2.5_ugm3"
]

input_df = pd.DataFrame([latest_data])[feature_columns]
scaled_input = scaler.transform(input_df)

# Generate 72-hour forecast
predictions = model.predict(scaled_input)[0]

# =========================================================
# 5. Build Result DataFrame
# =========================================================
start_time = datetime.now().replace(
    minute=0, second=0, microsecond=0
) + timedelta(hours=1)

forecast_df = pd.DataFrame({
    "Time": [start_time + timedelta(hours=i) for i in range(72)],
    "Predicted_AQI": predictions
})

print("\nðŸ“ˆ Next 72 Hours AQI Forecast:")
print(forecast_df)


```
