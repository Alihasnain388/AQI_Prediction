```python
import streamlit as st
import pandas as pd
import mlflow
import mlflow.sklearn
import joblib
import os
from pymongo import MongoClient
from datetime import datetime, timedelta
import plotly.express as px

# 1. Setup Environment (Map custom CI/CD secrets)
os.environ["MLFLOW_TRACKING_USERNAME"] = os.getenv("DAGSHUB_USERNAME", "Alihasnain388")
os.environ["MLFLOW_TRACKING_PASSWORD"] = os.getenv("DAGSHUB_TOKEN", "")
mlflow.set_tracking_uri("https://dagshub.com/Alihasnain388/AQI_Model.mlflow")

# 2. Page Configuration
st.set_page_config(page_title="Pearls AQI Predictor", page_icon="ðŸŒ¤ï¸")
st.title("ðŸŒ¤ï¸ Pearls AQI Predictor")
st.markdown("Predicting AQI for the next 72 hours using a 100% serverless stack.")

@st.cache_resource
def load_model_and_scaler():
    client = mlflow.tracking.MlflowClient()
    MODEL_NAME = "Karachi_AQI_Model"
    versions = client.search_model_versions(filter_string=f"name='{MODEL_NAME}'")
    latest = sorted(versions, key=lambda x: int(x.version), reverse=True)[0]
    
    model = mlflow.sklearn.load_model(f"models:/{MODEL_NAME}/{latest.version}")
    scaler_path = mlflow.artifacts.download_artifacts(run_id=latest.run_id, artifact_path="model/scaler.pkl")
    scaler = joblib.load(scaler_path)
    return model, scaler

@st.cache_data(ttl=3600) # Refresh every hour
def get_latest_features():
    mongo_uri = os.getenv("MONGO_URI", "your_mongodb_uri_here")
    mongo_client = MongoClient(mongo_uri)
    db = mongo_client["AQIPredictionSystem"]
    return db["karachi_features"].find_one(sort=[("timestamp", -1)])

# 3. Execution Logic
try:
    model, scaler = load_model_and_scaler()
    latest_data = get_latest_features()

    if latest_data:
        # Prepare Input
        feature_cols = ["hour", "day_of_week", "aqi_lag_1h", "aqi_lag_24h", "aqi_change_rate", "Wind_Speed_kmh", "PM2.5_ugm3"]
        input_df = pd.DataFrame([latest_data])[feature_cols]
        
        # Display Current Status
        st.subheader("ðŸ“ Current Air Quality Status")
        col1, col2, col3 = st.columns(3)
        col1.metric("Current PM2.5", f"{latest_data['PM2.5_ugm3']} Âµg/mÂ³")
        col2.metric("Wind Speed", f"{latest_data['Wind_Speed_kmh']} km/h")
        
        # Generate Forecast
        scaled_input = scaler.transform(input_df)
        predictions = model.predict(scaled_input)[0]
        
        start_time = datetime.now().replace(minute=0, second=0, microsecond=0) + timedelta(hours=1)
        forecast_df = pd.DataFrame({
            "Time": [start_time + timedelta(hours=i) for i in range(72)],
            "Predicted_AQI": predictions
        })

        # Display Forecast Chart
        st.subheader("ðŸ“ˆ 72-Hour AQI Forecast")
        fig = px.line(forecast_df, x="Time", y="Predicted_AQI", title="Predicted AQI Trend")
        st.plotly_chart(fig, use_container_width=True)

        # Show Table
        with st.expander("View Raw Forecast Data"):
            st.dataframe(forecast_df)
    else:
        st.error("No feature data found in MongoDB.")

except Exception as e:
    st.error(f"Error loading dashboard: {e}")

st.divider()
st.caption("Powered by GitHub Actions, DagsHub MLflow, and MongoDB Feature Store.")
```
